###5. What is Information Gain, and how is it used in Decision Trees

Information gain claculates, how much impurity is reduced after a node is splitted.
It compares the entropy of the parent node(before split) with the  weighted average of the entropy of the child nodes(after split).
Higher information gain indicates more effective split.

###6. What is the difference between Gini Impurity and Entropy

Gini impurity - Calculates the probability of misclassification,the output value ranges between 0 - 0.5

Entropy- Calculates the uncertainty within a data set. output value ranges between 0 to log2n where n is the number of classes


###7. What is the mathematical explanation behind Decision Trees
To reduce the impurity step by step until max depth/ min leaf criteria is reached or all the leaf nodes become pure.
Gini/ entropy is used to measure the purity of a node. For each node all possible splittin option is compared based on impurity to decide the best possible one.


###8. What is Pre-Pruning in Decision Trees
Restrict the growth of decision tree, to avoid overfitting.
Pre pruning is done before the entire tree formation is completed using max depth, impurity thresold,minimum sample per node tec.


###9. What is Post-Pruning in Decision Trees
First the entire tree is constructed that achieves purity for each node. the the sub trees are evaluated to see whether removing any of those can make the tree generalized.

Post pruning can be done using hyper parameter tuning, after the entire tree is constructed, to find the best parameters to restrict the tree.


###10. What is the difference between Pre-Pruning and Post-Pruning
Pre pruning - Stops the tree growth based on parameters.Reduces the trainng time.

Post pruning - Prunes a fully grown tree. removes branches or nodes that add complexity without improving.Simplifies the model while retaining useful nodes.


###11. What is a Decision Tree Regressor
Predicts continuous numeric values. The output of each node is the mean of the target variable.Recursive partitioning is done based on MSE/MAE to split the nodes and minimize the error.


###12. What are the advantages and disadvantages of Decision Trees
  - Advantages
    - Simple to interpret
    - Works well with large data sets
    - Used for both classification & regression tasks
  
  -Disadvantages
    - greedy algorithm tries to momorise the data, that leads to overfitting
    - Lower accuracy compared to ensemble models
    - Feature with high number of unique values dominate the split, leading to biased prediction


###13. How does a Decision Tree handle missing values

###14. How does a Decision Tree handle categorical features
Decision tree considers each category as a distinct value. Split is decided based on the information gain to find the optimal criteria.

```python
###5.Write a Python program to train a Decision Tree Classifier and visualize the tree using graphviz

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
import graphviz
from sklearn import tree

warnings.filterwarnings('ignore')

iris = load_iris()
X,y = load_iris(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = DecisionTreeClassifier(criterion='gini')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_pred,y_test)
print(f"Accuracy {accuracy}")

dot_data = export_graphviz(model, out_file=None,
                           feature_names=iris.feature_names,
                           class_names=iris.target_names,
                           filled=True, rounded=True,
                           special_characters=True)

# Render the tree
graph = graphviz.Source(dot_data)
graph.render("decision_tree")
graph.view()

tree.plot_tree(model,filled=True)
plt.show()
```

```python
###10.  Write a Python program to train a Decision Tree Classifier and display the feature importance scores

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

warnings.filterwarnings('ignore')

iris = load_iris()
X,y = load_iris(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_pred,y_test)
print(f"Accuracy {accuracy}")
print(f"Feature importance : {model.feature_importances_}")

# Display feature importance scores
feature_importances = model.feature_importances_
importance_df = pd.DataFrame({'Feature': iris.feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print("\nFeature Importance Scores:")
print(importance_df)

# Visualize feature importance using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
###12.  Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and
##visualize its effect on accuracy

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

path = model.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

train_accuracies = []
test_accuracies = []

for alpha in ccp_alphas:
    pruned_model = DecisionTreeClassifier(ccp_alpha=alpha)
    pruned_model.fit(X_train, y_train)
    train_accuracies.append(accuracy_score(y_train, pruned_model.predict(X_train)))
    test_accuracies.append(accuracy_score(y_test, pruned_model.predict(X_test)))

# Visualize the effect of CCP on accuracy
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_accuracies, marker='o', label="Train Accuracy")
plt.plot(ccp_alphas, test_accuracies, marker='o', label="Test Accuracy")
plt.xlabel("CCP Alpha")
plt.ylabel("Accuracy")
plt.title("Effect of Cost Complexity Pruning (CCP) on Accuracy")
plt.legend()
plt.grid()
plt.show()
```

```python
###13. Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision,
###Recall, and F1-Score

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score

warnings.filterwarnings('ignore')


X,y = load_iris(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
precision = precision_score(y_pred,y_test, average='weighted')
recall = recall_score(y_pred,y_test, average='weighted')
f1 = f1_score(y_pred,y_test, average='weighted')
print(f"Precision- {precision} Recall- {recall} F1- {f1}")
```

```python
###15. Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values
###for max_depth and min_samples_split

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

warnings.filterwarnings('ignore')


X,y = load_iris(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)

model = DecisionTreeClassifier()
params = {
    "max_depth":[1,3,5,10,20],
    "min_samples_split": [1,2,3,4,5]
    }

model = GridSearchCV(param_grid=params, cv=5, verbose=0,estimator =model)
model.fit(X_train,y_train)

estimator = model.best_estimator_
print(f"Bets parameter: {model.best_params_}")
y_pred = estimator.predict(X_test)
accuracy = accuracy_score(y_pred,y_test)
print(f"Accuracy {accuracy}")
```
