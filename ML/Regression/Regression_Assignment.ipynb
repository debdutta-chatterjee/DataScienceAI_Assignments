###1. What is Simple Linear Regression?

Statistical method to predict 1 dependent variable based on independent variables. The goal is to find the linear equation or best fit line Y = mX +C to predict the dependent variable values.


###2. What are the key assumptions of Simple Linear Regression?

Linearity - The relationship between dependent(Y) & independent (X) variable nmust be linear

Independence - The error from one observation is not related to other observations

Homoscedasticity - The residual variance is constant

Normality - The residuals are normally distributed

No perfect multicollinearity - The independent variables are not perfectly correlated to each other

```python
###31  How is polynomial regression implemented in Python?

##import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#Generate X
np.random.seed(1)
X= np.random.rand(100,1)

#Generate Y
y = 5+ 2*X+ 6*X**2+np.random.rand(100,1)

#Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
print("Data -",X_train.shape,X_test.shape,y_train.shape,y_test.shape)

#Model fit
poly_features = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly_features.fit_transform(X_train)
model = LinearRegression()
model.fit(X_train_poly,y_train)
print(model.coef_, model.intercept_)

#Prediction on train data
y_pred = model.predict(X_train_poly)
#MSE
mse = mean_squared_error(y_train,y_pred)
print(mse)
r2 = r2_score(y_train, y_pred)
print(r2)



#Prediction on test data
X_test_poly = poly_features.transform(X_test)
y_pred = model.predict(X_test_poly)
#MSE
mse = mean_squared_error(y_test,y_pred)
print(mse)
r2 = r2_score(y_test, y_pred)
print(r2)


# Plotting
plt.scatter(X_test, y_test, color='red', label='Test data')
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
plt.plot(X_range,model.predict(poly_features.transform(X_range)), color='red', label='Polynomial regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
