###1. What is hypothesis testing in statistics

A fundamental method in statustucs to verify that whether is enough evidence in a sample of data about certain condition holds true for the entire population

###2. What is the null hypothesis, and how does it differ from the alternative hypothesis

Null hypothesis is the default assumption that the researchers try to disprove. it has the equality symbol.
Example - H0 -> The average age of the people in a city is 50 years

Alternate hypothesis is the opposite to the null hypothesis, that the researchers try to prove.
Example HA -> The average age of the people in a city is not 50

###3. What is the significance level in hypothesis testing, and why is it important

Significance level α is the thresld set by the researchers, to represent the probability of rejecting the null hypothesis when its actually true. It is the criteria, to decide whether the data is sufficiently inconsistent with the null hypothesis assumption.

It is important to perfrom decision making, error control.

Example - If we get the p value from a hypothesis testing as 0.04 & the α is 0.05. then we reject the null hypothesis.


###4. What does a P-value represent in hypothesis testing?
p value is the probability of the test statistics.It quantifies how likely it is to see the observed data, or something more extreme, purely by random chance if the null hypothesis is correct.

###5. How do you interpret the P-value in hypothesis testing?

If the p value is small, then we reject the null hypothesis.
If the p value is large, then we fail to reject the null hypothesis.
Example - If we get the p value from a hypothesis testing as 0.04 & the α is 0.05. then we reject the null hypothesis.
if the p value is 0.06 then we fail to reject the null hypothesis.

###6. What are Type 1 and Type 2 errors in hypothesis testing?
Type 1 error - False positive - Reject the null hypothesis when it is true.

Type 2 error - False Negative - Fail to reject the null hypothesis when it is false.

###7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One tail tests identifies the probability of a tests statistics falling in one side of the rejection area in the distribution.
Example H0 = μage >= 30

Two tail test identifies the probability of a tests statistics falling in both side of the rejection area in the distribution.
Example H0 = μage =/= 30

```python
###1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
##interpret the results

import numpy as np
from scipy import stats

# Sample data
sample_data = [12, 15, 14, 10, 13, 14, 11, 15, 13, 13]
population_mean = 14 # Lets assume
alpha = 0.05

# Calculating sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Performing the Z-test
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = stats.norm.sf(abs(z_score)) * 2

# Interpretation
print(f'Sample Mean: {sample_mean}')
print(f'Z-score: {z_score}')
print(f'P-value: {p_value}')

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
###5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def type_1_2_errors(alpha, sample_mean, population_mean, population_std, n):
    z_critical = stats.norm.ppf(1 - alpha / 2)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    #Type II error probability
    power = stats.norm.sf((z_critical - z_score)) + stats.norm.cdf((-z_critical - z_score))
    beta = 1 - power

    # Plotting the decision regions and the Z-distributions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(12, 8))

    # Plot standard normal distribution (null hypothesis)
    plt.plot(x, y, label='Null Hypothesis (H0)')
    plt.fill_between(x, y, where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.3, label='Type I Error (α)')

    y_alt = stats.norm.pdf(x, loc=z_score)
    plt.plot(x, y_alt, label='Alternative Hypothesis (H1)')
    plt.fill_between(x, y_alt, where=(x < z_critical) & (x > -z_critical), color='blue', alpha=0.3, label='Type II Error (β)')
    plt.axvline(x=z_critical, color='black', linestyle='--')
    plt.axvline(x=-z_critical, color='black', linestyle='--')
    plt.axvline(x=z_score, color='green', linestyle='--', label=f'Z-score: {z_score:.2f}')
    plt.title(f'Type I and Type II Errors (α={alpha}, β={beta:.2f}, Power={power:.2f})')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()


alpha = 0.05
sample_mean = 13
population_mean = 14
population_std = 2
n = 30
type_1_2_errors(alpha, sample_mean, population_mean, population_std, n)
```

```python
###9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    # Calculate the confidence interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [12, 15, 14, 10, 13, 14, 11, 15, 13, 13]
confidence_level = 0.95
ci_lower, ci_upper = confidence_interval(sample_data, confidence_level)
print(f'Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})')
```

```python
###10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    return margin_of_error

sample_data = [12, 15, 14, 10, 13, 14, 11, 15, 13, 13]
confidence_level = 0.95
margin_of_error = confidence_interval(sample_data, confidence_level)
print(f'Margin of error: ({margin_of_error:.2f})')
```

```python
###11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayesian_inference(prior, likelihood, positive):
    p = (likelihood * prior) / positive
    return p

#Known probability of having the disease
Known_probability = 0.01  #1% of the population can have the disease

#Probability of a positive test result given the disease
likelihood_positive_test_result = 0.99

#Probability of a positive test result given no disease
likelihood_positive_test_no_disease = 0.05

#Total probability of a positive test result
positive = (likelihood_positive_test_result * Known_probability) + \
                               (likelihood_positive_test_no_disease * (1 - Known_probability))

#Calculate the probability of having the disease given a positive test result
probability = bayesian_inference(Known_probability, likelihood_positive_test_result, positive)
print(f'Probability: {probability:.4f}')
```

```python
###20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
##results

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Method': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'A', 'A', 'A', 'B', 'B', 'B'],
    'Gender': ['M', 'M', 'M', 'F', 'F', 'F', 'M', 'M', 'M', 'F', 'F', 'F', 'M', 'M', 'M'],
    'Score': [85, 88, 90, 83, 86, 88, 80, 82, 85, 84, 87, 89, 81, 84, 87]
}

df = pd.DataFrame(data)

model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=df).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

sns.pointplot(x='Method', y='Score', hue='Gender', data=df, markers=['o', 's'], linestyle='-')
plt.title('Interaction Plot of Teaching Method and Gender')
plt.xlabel('Teaching Method')
plt.ylabel('Score')
plt.show()
```

```python
###22.Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Method': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Score': [85, 88, 90, 83, 86, 88, 80, 82, 85]
}

df = pd.DataFrame(data)

fvalue, pvalue = stats.f_oneway(
    df[df['Method'] == 'A']['Score'],
    df[df['Method'] == 'B']['Score'],
    df[df['Method'] == 'C']['Score']
)

sns.boxplot(x='Method', y='Score', data=df)
plt.title('Boxplot of Scores by Teaching Method')
plt.xlabel('Teaching Method')
plt.ylabel('Score')
plt.show()
```

```python
###23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

data1 = np.random.normal(50, 5, 100)
data2 = np.random.normal(55, 5, 100)

plt.hist(data1, bins=20, label='Group 1: Mean = 50, SD = 5')
plt.hist(data2, bins=20, label='Group 2: Mean = 55, SD = 5')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

t_stat, p_value = stats.ttest_ind(data1, data2)
print(f'T-statistic: {t_stat}')
print("p-value:", p_value)
```

```python
###24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy import stats

group_a = np.random.normal(loc=50, scale=10, size=30)
sample_variance = np.var(group_a, ddof=1)
n = len(group_a)

pop_variance = 100 #H0

chi_square_stat = (n - 1) * sample_variance / pop_variance
print("Chi-square statistic:", chi_square_stat)

critical_value_low = stats.chi2.ppf(0.025, df=n-1)
critical_value_high = stats.chi2.ppf(0.975, df=n-1)
print("Critical value (low):", critical_value_low)
print("Critical value (high):", critical_value_high)

if chi_square_stat < critical_value_low or chi_square_stat > critical_value_high:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
###25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

count_group1 = 50  # Number of successes in Group 1
nobs_group1 = 100  # Total observations in Group 1

count_group2 = 60  # Number of successes in Group 2
nobs_group2 = 120  # Total observations in Group 2

count = np.array([count_group1, count_group2])
nobs = np.array([nobs_group1, nobs_group2])

z_stat, p_value = proportions_ztest(count, nobs)

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

```python
###26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
from scipy import stats

np.random.seed(1)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=20, scale=8, size=30)
group1_var = np.var(group1, ddof=1)
group2_var = np.var(group2, ddof=1)

f_stat = group1_var / group2_var
dof_group1 = len(group1) - 1
dof_group2 = len(group2) - 1
p_value = 1 - stats.f.cdf(f_stat, dof_group1, dof_group2)

p_value *= 2

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

plt.hist(group1, alpha=0.5, label='Dataset 1', bins=10)
plt.hist(group2, alpha=0.5, label='Dataset 2', bins=10)

plt.axvline(np.mean(group1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(group2), color='orange', linestyle='dashed', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.title('Histograms of Dataset 1 and Dataset 2')
plt.show()
```

```python
###27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
from scipy import stats

observed = np.array([50, 30, 20])
total_sum = observed.sum()
expected = np.array([total_sum / 3, total_sum / 3, total_sum / 3])

print("Observed frequencies:", observed)
print("Expected frequencies:", expected)

chi_square_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_square_stat)
print("p-value:", p_value)
```
