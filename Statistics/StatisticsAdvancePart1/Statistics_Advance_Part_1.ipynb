````python
###9.  Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_score(data):
  mean = np.mean(data)
  std = np.std(data)

  z_score = [(x-mean)/std for x in data]
  return z_score

def plot_z_score(z_score):
  sns.histplot(z_score,bins=50,kde= True)
  plt.title("Z Score Histogram")
  plt.xlabel("Z Score")
  plt.ylabel("Density")
  plt.show()

normal = np.random.normal(100,10,10000)
plot_z_score(calculate_z_score(normal))

````

````python
###10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Poisson distribution
lambda_value = 5
num_samples = 1000
sample_sizes = [5, 10, 30, 50, 100]

# Generate data from a Poisson distribution
data = np.random.poisson(lam=lambda_value, size=num_samples)

# Plot histograms of sample means for different sample sizes
for i, sample_size in enumerate(sample_sizes, 1):
    # Draw random samples and calculate their means
    sample_means = [np.mean(np.random.choice(data, sample_size)) for _ in range(num_samples)]

    # Plot histogram of sample means
    plt.subplot(2, 3, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g', label='Sample Means')

    # Fit a normal distribution to the sample means
    mu, std = norm.fit(sample_means)

    # Plot the fitted normal distribution
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)

    plt.plot(x, p, 'k', linewidth=2)
    plt.title(f'Sample Size = {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
````

````python
###19. Implement hypothesis testing using Z-statistics for a sample dataset


import numpy as np
from scipy.stats import norm

#Sample data
sample_data = [12, 15, 14, 10, 13, 14, 16, 15, 13, 14]
population_mean = 13
population_std = 2
significance_level = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)
sample_size = len(sample_data)

#Z Stats
z_statistic = (sample_mean - population_mean) / (population_std  / np.sqrt(sample_size))

critical_value = norm.ppf(1 - significance_level / 2)
print(f"Critical value for significance level {significance_level}: ¬±{critical_value:.2f}")

# Hypothesis testing decision
if abs(z_statistic) > critical_value:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
````

````python
###21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy import stats

# Generate random data from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))
    h = std_error * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return mean, mean - h, mean + h

# Calculate the 95% confidence interval
mean, lower_bound, upper_bound = confidence_interval(data)

print(f"Generated Data Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
````

````python
###24. Simulate a random variable using a continuous uniform distribution and calculate its expected value

import numpy as np
import scipy.integrate as integrate

a =10
b = 20

size = 1000  # number of samples

# Generate data from the uniform distribution
data = np.random.uniform(a, b, size)


# Define the PDF of the uniform distribution
def uniform_pdf(x, a, b):
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

# Integrate to calculate the expected value
expected_value, _ = integrate.quad(lambda x: x * uniform_pdf(x, a, b), a, b)

print(f"Expected Value calculated using the integral formula: {expected_value:.2f}")
print(f"Expected Value calculated using the mean: {np.mean(data):.2f}")

````

````python
###27. Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Generate data from a normal distribution
mean = 50
std_dev = 10
num_samples = 1000
data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Perform Z-score normalization
normalized_data = z_score_normalization(data)

# Plot the original data and normalized data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7, label='Original Data')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7, label='Normalized Data')
plt.title('Z-score Normalized Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
````
